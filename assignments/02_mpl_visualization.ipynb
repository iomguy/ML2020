## Генерация данных с помощью библиотеки numpy. Элементарные функции. 

На следующем занятии мы будем учиться обрабатывать реальные данные разного типа из многомерных таблиц, но на данном этапе более удобным и полезным будет попробовать симулировать и визуализировать данные на основе простых числовых массивов. Для этого удобно использовать библиотеку numpy: некоторые методы генерации массивов приведены ниже, чтобы посмотреть еще опции, можно заглянуть в [документацию](https://numpy.org/doc/1.19/reference/routines.array-creation.html). 

Некоторые способы генерировать массивы чисел: 
*   `np.linspace(start,stop,num)` - генерация массива чисел размера num, равномерно распределенных на интервале `[start,stop]`. 
*   `np.logspace` - то же, что `np.linspace`, но на логарифмической шкале. 
*   `np.arange(start,stop,step)` - генерация массива чисел в диапазоне `[start,stop]`, где каждый следующий элемент на step больше предыдущего.
*   `np.zeros(shape), np.ones(shape) `- возвращает массив заданной формы shape, заполненный нулями и единицами соответственно. 
*  `np.empty(shape) `- пустой массив заданной формы.
*  `np.eye(shape)` - массив, заполненный нулями, с диагональю, заполненной единицами.

Конвертация из объекта данных: 
*   `np.array(a)`
*   `np.asarray(a)` - конвертирует входные данные a в массив numpy.
*   np.asmatrix(data) - конвертирует входные данные data в матрицу numpy









## Инициализация изображения. Расположение графиков. 

Figure - это общее изображение, которое будет отображаться в результате запуска кодовой ячейки, пространство для размещения графиков. Сами графики, любое количество которых можно разместить в рамках одной фигуры, называются axes. Фигуру и графики можно задавать разными способами.  

1. Задавать сначала фигуру как отдельный объект, а затем изменять количество графиков внутри вручную: например, добавлять их с помощью функций `add_axes` и `add_subplot`. В большинстве случаев удобнее использовать `add_subplot(n,m,i)`, где `n,m` - это количество графиков в фигуре по вертикали и горизонтали соответственно (`n*m` - размер сетки графиков), а `i` - номер текущего графика в сетке. В `add_axes `расположение графика определяется 4 вещественными числами: `[left,bottom,width,height]`. В случае инициализации фигуры таким способом обращение к ее внутренним графикам происходит по имени каждого из них. 

```
fig=plt.figure(figsize=(10,3))
ax_1=fig.add_subplot(131)
ax_2=fig.add_subplot(132)
ax_3=fig.add_subplot(133)
ax_1.set_title("Squared X")
ax_1.plot(x,x**2)
ax_2.set_title("Logarithm of X")
ax_2.plot(x,np.log(x))
ax_3.set_title("Square root of X")
ax_3.plot(x,np.sqrt(x))
fig.tight_layout() #полезная функция, автоматически выравнивает расстояние между графиками, чтобы они не наезжали друг на друга
fig.show()
```

В этом случае обращение к отдельным графикам происходит по индексу их местонахождения в сетке, и в целом взаимодействие с изображением строится как с единым объектом, а не с отдельными сущностями.

## Оформление графиков

Какую бы симпатичную сетку графиков мы ни задали изначально, визуализация данных не будет иметь никакого смысла, если изображение и его составные части не подписаны и не обозначены адекватным образом. Чтобы по картинке можно было понять, что именно происходит в данных на этом этапе и почему, она должна иметь общее название, подписи координатных осей, легенду (если, например, на одном графике находятся сразу несколько линий, соответствующих разным переменным). Ниже перечислены основные функции для оформления картинок. 

*  ` plt.title(), ax.set_title()` - название графика
*   `plt.xlabel(),plt.ylabel(), ax.set_xlabel(), ax.set_ylabel()` - наименования осей
* `plt.legend()` - выведение легенды графика на экран; наименования графиков в легенде задаются в функции построения самой картинки (`plt.plot()` в параметре `"label"`
*   `plt.xticks(),plt.yticks(), ax.set_xticks(), ax.set_yticks()` - для категориальных перменных; расположение и наименования категорий, отложенных на осях.
*  `plt.xlim(), plt.ylim(), ax.set_xlim(), ax.set_ylim() `- левые и правые границы значений, отложенных на соответствующих осях.





```
x=np.linspace(0.1,3.0,num=1000)
y_1,y_2,y_3,y_4=4**x,np.log10(x), x**x, np.sqrt(x)
plt.plot(x,y_1,'-k',label='y=4^x')
plt.plot(x,y_2,'-r',label='y=log10(x)')
plt.plot(x,y_3,'-g',label='y=x^x')
plt.plot(x,y_4,'-b',label='y=sqrt(x)')
plt.title("Сравнение роста скорости элементарных функций на участке x=[0.1,3.0]")
plt.xlabel("Ось x")
plt.ylabel("Ось y")
plt.legend()
plt.show()
```

```
fig,ax=plt.subplots(1,3)
fig.set_size_inches(9,3)
ax[0].plot(x**(1/3),'-r')
ax[0].set_title("График функции x^(1/3)")
ax[0].set_xlabel("Ось x")
ax[0].set_ylabel("Ось y")
ax[1].plot(2**(1/x),'-y')
ax[1].set_title("График функции 2^(1/x)")
ax[1].set_xlabel("Ось x")
ax[1].set_ylabel("Ось y")
ax[2].plot(np.log2(x))
ax[2].set_title("График функции log2(x)")
ax[2].set_xlabel("Ось x")
ax[2].set_ylabel("Ось y")
fig.tight_layout()
fig.show()
```

## Ключевые характеристики при построении графиков. Типы графиков

Для каждого вида графика, помимо данных, которыми задаются координатные оси (x,y), существует много дополнительных параметров, настройка которых существенно влияет на то, как будет восприниматься информация на изображении. Они варьируются в зависимости от конкретной функции, но можно выделить несколько общих ключевых черт: форма отображения данных на графике (точка, линия, столбик, и т.д.), цвет, размер, используемая палитра оттенков. Кроме того, значения, откладываемые по координатным осям, могут нуждаться в дополнительных преобразованиях, чтобы график был более информативен.

### График `plt.plot()`

Базовый график библиотеки matplotlib, позволяющий отобразить зависимость между двумя и более переменными в декартовой системе координат. 
Эта функция содержит большое число аргументов, рассмотрим несколько основных. 


1. ` x, y` - данные, откладываемые на соответствующих координатных осях.
2. `[fmt]`, или, в расшифровке `[marker][line][color]` - это строка-комбинация коротких символов, обозначающих тип используемой при построении графика точки, линии и их цвет. 




| Marker|  | Line |    | Color |   |
| --- | --- | --- | --- |--- | --- |
| '**.**'| point marker|'**-**'| solid line style| '**b**' | blue|
| '**,**'| pixel marker | '**--**' | dashed line style  |'**g**' | green  |
| '**o**'| circle marker  | '**-.**' | dash-dot line style  |'**r**' | red |
| '**v**'| triangle_down marker | '**:**' | dotted line style  |'**c**' | cyan  |
| '**^**'| triangle_up marker  |  |   |'**m**'| magenta  |
| '**<**'| triangle_left marker  |  |   |'**y**' | yellow |
| '**>**'| triangle_right marker  |  |   |'**k**' |  black |
| '**s**'| square marker |  |   |'**w**' | white |
| '**p**'| pentagon marker | |   | |   |
| '*'| star marker |  |   | |   |
| '**+**'| plus marker |  |   | |   |


3. `data` - индексированный объект, содержащий размеченные данные (например, словарь или таблица). Если визуализируется содержимое объекта такого рода, то в аргументах `x,y` указываются метки/названия соответствующих переменных. 
Кроме того, в данной функции есть большой набор дополнительных параметров, позволяющих, например, отрегулировать ширину линии (`linewidth`), окрашивать границы точек (`markeredgecolor`), регулировать их размер (`markersize`), и т.д. Подробный список всех настраиваемых характеристик графика можно изучить [здесь](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.plot.html).

```
fig,ax=plt.subplots(2,2)
fig.set_size_inches(7,5)
ax[0,0].plot(x[:50],np.sin(x[:50]),'--g')
ax[0,1].plot(x[:50],np.cos(x[:50]),'o--m')
ax[1,0].plot(x,np.sqrt(x),'-.b')
ax[1,1].plot(x,1/x,'-k')
fig.tight_layout()
fig.show()
```

### Диаграмма рассеяния, или Scatterplot. График `plt.scatter()`

Диаграмма рассеяния- близкий родственник классического lineplot, который уже рассматривался выше, с той только разницей, что вместо линий здесь фигурируют точки. Каждая точка на диаграмме рассеяния представляет собой один объект в данных, по осям отложены значения двух интересующих признаков. Зачем нужен scatterplot? Построение этого графика полезно, когда мы обращаемся к данным как к совокупности отдельных объектов: тогда, помимо признаков, отложенных на координатных осях, характеристики точек на графике (цвет, размер, форма, прозрачность) тоже могут отражать набор определенных признаков. Это позволяет увидеть определенные зависимости и структурные особенности в данных в многомерном пространстве. 


```
x_1=np.linspace(0,100)
plt.plot(x_1,x_1**2,'-r',label='y=x^2')
plt.scatter(x_1,x_1**2+[np.random.randint(-800,800) for i in range(50)],c='g',alpha=0.7,edgecolors='k',label='y=x^2+шум в данных')
```

```
from sklearn.datasets import load_iris
iris=load_iris()
x_index=0
y_index=1
plt.figure(figsize=(5, 4))
formatter = plt.FuncFormatter(lambda i, *args: iris.target_names[int(i)])
plt.scatter(iris.data[:, x_index], iris.data[:, y_index], c=iris.target, edgecolors='k',alpha=0.7)
plt.colorbar(ticks=[0, 1, 2],format=formatter)
plt.xlabel(iris.feature_names[x_index])
plt.ylabel(iris.feature_names[y_index])
plt.title('Диаграмма рассеяния для данных о 3 видах ириса')
plt.tight_layout()
plt.show()
```

### Распределения и гистограммы. График `plt.hist()`

Построение гистограммы помогает получить представление о распределении данных, оценить количество выбросов, среднее значение признака и его дисперсию. По горизонтальной оси в гистограмме отложены интервалы возможных значений признака, по вертикали - частота встречаемости текущего значения в данных. Основной регулируемый параметр в гистограмме - интервалы (`bins`): их количество, область допустимых значений, и т.д. Кроме того, можно настроить тип гистограммы (`histtype`), расположение интервалов (`align, orientation`), ширину столбиков (`rwidth`), цвет (`color`), и другие характеристики, подробно описанные [здесь](https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.hist.html). Ниже приведены примеры гистограмм для 500 случайных чисел, сгенерированных из различных распределений. 

```
fig,ax=plt.subplots(1,3)
fig.set_size_inches(10,3)
ax[0].hist(np.random.standard_normal(size=500),bins=50)
ax[0].set_title("Histogram of stand.normal distribution")
ax[0].set_xlabel("Randomly generated samples")
ax[0].set_ylabel("Frequency")
ax[1].hist(np.random.exponential(size=500),bins=100,color='green')
ax[1].set_title("Histogram of exponential distribution")
ax[1].set_xlabel("Randomly generated samples")
ax[1].set_ylabel("Frequency")
ax[2].hist(np.random.poisson(size=500),bins=75,color='red')
ax[2].set_title("Histogram of Poisson distribution")
ax[2].set_xlabel("Randomly generated samples")
ax[2].set_ylabel("Frequency")
fig.tight_layout()
fig.show()
```

### Столбцовая диаграмма, или bar plot. Графики `plt.bar(), plt.barh()`

Основное отличие столбцовой диаграммы от гистограммы - гистограмма работает с данными непрерывного типа, группируя их в настраиваемое и регулируемое число интервалов; столбцовая диаграмма строится для сравнения категориальных/ранжированных переменных, и количество столбиков в ней строго соответствует числу категорий в данных. В зависимости от ориентации - горизонтальной и вертикальной - по одной оси откладываются исследуемые категории, по другой - значения признака, по которому они сравниваются. Два основных параметра при постройке bar plot - `x, height`: `x` - это координаты столбцов на числовой оси, а `height` - их высота, которой является значение измеряемого признака. 

```
bars_dict={'WöD':9.8,'Lure':9.7,'Хроники':6.0,'Гусары':10.0,'Ossi':8.5,'Мачты':8.8,'Mishka':7.8,'Brimborium':8.0,'Залив':5.7}
ratings=sorted(bars_dict.values(),reverse=True)
objects=[list(bars_dict.keys())[list(bars_dict.values()).index(i)] for i in ratings]
y_pos=np.arange(len(objects))
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.bar(y_pos, ratings, align='center', alpha=0.8,color='g')
plt.xticks(y_pos, objects,rotation=45)
plt.ylabel('Рейтинг')
plt.title('Рейтинги петербургских баров')

plt.subplot(122)
plt.barh(y_pos, ratings, align='center', alpha=0.8,color='g')
plt.yticks(y_pos, objects)
plt.xlabel('Рейтинг')
plt.title('Рейтинги петербургских баров')
plt.tight_layout()
plt.show()
```

### Тепловая карта, или heatmap. График `plt.imshow()`

Тепловая карта - это графическое представление таблицы данных, где величина значения в таблице соответствует интенсивности цвета. Самый простой пример использования тепловой карты - при построении корреляционной матрицы: чем сильнее связь между объектами, тем теплее и ярче будет цвет соответствующей ячейки. При построении heatmap важно использовать правильную цветовую шкалу, подробнее об этом рассказано в [лекции](https://www.youtube.com/watch?v=O2zXpQmij_c).

```
corr_map=np.random.random_sample((50,50))
a,b,c,d=np.random.randint(10,20),np.random.randint(25,30),np.random.randint(40,45),np.random.randint(50,55)
corr_map[a:b,c:d]=corr_map[a:b,c:d]*5
plt.figure(figsize=(10,5))
plt.subplot(131)
plt.imshow(corr_map)
plt.subplot(132)
plt.imshow(corr_map,cmap='gray',interpolation='bilinear')
plt.subplot(133)
plt.imshow(corr_map,cmap='bwr',interpolation='bicubic')
```
